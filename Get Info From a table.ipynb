```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url='https://companiesmarketcap.com/largest-companies-by-revenue/'
```

```python
data=requests.get(url).text
```

```python
soup=BeautifulSoup(data,'html.parser')
```

```python
tables=soup.find_all('table')
for i,x in enumerate(tables):
    index=i
print(index)
```

```python
companies=pd.DataFrame(columns=['Name','Revenue (Billions)','Price','Country'])
for row in tables[0].tbody.find_all('tr'):
    col=row.find_all('td')
    if (col!=[]):
        Name=col[1].text
        Revenue=col[2].text
        Price=col[3].text
        Country=col[6].text
        companies=companies.append({'Name':Name,'Revenue (Billions)':Revenue,'Price':Price,'Country':Country},ignore_index=True)
```

```python
companies
```

```python
companies['Name'] = companies['Name'].str.replace('\n\n \n\n', "")
companies['Name'] = companies['Name'].str.replace('\n \n', "")
companies['Name'] = companies['Name'].str.replace('\n', "-")
companies
```

```python
companies['Revenue (Billions)']=companies['Revenue (Billions)'].astype(str)
companies['Revenue (Billions)']=companies['Revenue (Billions)'].str.replace(" B \n","")
companies['Revenue (Billions)']=companies['Revenue (Billions)'].str.replace(" B  \n","")
companies['Revenue (Billions)']=companies['Revenue (Billions)'].str.replace("$","").astype(float)
```

```python
companies['Price']=companies['Price'].astype(str)
companies['Price']=companies['Price'].str.replace("$","").astype(float)
```

```python
companies
```
